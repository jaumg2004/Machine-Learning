```python
inertia = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i, n_init='auto')
    kmeans.fit(clust1)
    inertia.append(kmeans.inertia_)
plt.plot(range(1, 8), inertia)
plt.title('Elbow')
```
